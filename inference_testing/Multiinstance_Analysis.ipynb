```python
import pandas as pd
import numpy as np
import os, cv2
import PIL
import matplotlib.pyplot as plt
```

```python
def read_image_grayscale(path):
    return cv2.imread(path, 0)
```

```python
MODEL_NAME = "11FCMN8rmsprop_lr0.0001_prep_keras_dp0.001_ep200.h5"
MODEL_PATH = os.path.join('..' , 'output', 'models', MODEL_NAME)
IMAGES_PATH = 'D:\\temp\\dataset_resize\\images_resize'
MASK_PATH = "D:\\temp\\dataset_resize\\masks_resize"
CSV_PATH = "C:\\Users\\Jorge\\Documents\\repos\\FCNBudDetection\\inference_testing\\multi_instance_dataset.csv"
PREDICTIONS_PATH = ".\\multi_predictions"
csv = pd.read_csv(CSV_PATH, index_col =False)
```

```python
mask_list = csv['mask'].tolist()
img_list = csv['imageOrigin'].tolist()
```

```python
def plot_images(img_index):
    rows, cols = 4, 4
    fig, ax = plt.subplots(rows, cols)
    img_index = img_index
    for row in range(rows):
        for col in range(cols):
            if np.remainder(row+col, 2) == 0:
                ax[row, col].imshow(read_image_grayscale(os.path.join(PREDICTIONS_PATH, "norm_pred_" + img_list[img_index])))
                ax[row, col].figure.set_size_inches(13,13)
                ax[row, col].set_title(img_list[img_index])
            else: 
                ax[row, col].imshow(read_image_grayscale(os.path.join(MASK_PATH, mask_list[img_index])))
                ax[row, col].figure.set_size_inches(13,13)
                ax[row, col].set_title(mask_list[img_index])
                img_index += 1

    plt.show()
```

```python
plot_images(0)
```
